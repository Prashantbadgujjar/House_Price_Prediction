# 7. Data Visualization

```python
# countplot for categorical column
sns.countplot(x='ocean_proximity',data=df)
plt.show()
```

```python
#target coulmn distribution

plt.figure(figsize=(10,5))
sns.histplot(x=df['median_house_value'], data=df, kde=True, bins=40)
plt.title('Target Column Distribution')
plt.xlabel('Median House Value')
plt.show()
```

```python
# Check outliers in each column

fig,axes = plt.subplots(3,3, figsize=(15,8))
axes=axes.flatten()
for i,col in enumerate(num_cols):
    sns.boxplot(x=df[col], ax=axes[i])
plt.tight_layout()
plt.show()
```

```python
# identigy presence of highly correlated and feature relationships
plt.figure(figsize=(20,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# correlation with target
corr_with_target=df[num_cols].corr()['median_house_value'].sort_values(ascending=False)
corr_with_target
```
